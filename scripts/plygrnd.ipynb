```python
from pathlib import Path
from overwrite_classes_from_cs_html import extract_courses_from_html
import os
import re
import random

HTML_PATH = Path(os.environ.get("CS_CLASSES_HTML", "CS classes.html"))
exp = r'(\w*)(\d{4})(\w*)'

courses = extract_courses_from_html(HTML_PATH)
courses = [[c['course_code'], c['credits']] for c in courses]
```

```python
def mode1_lowest_first(courses):
    for idx, c in enumerate(courses):
        match = re.findall(exp, c[0])
        courses[idx] = [match[0][1], match[0][0], c[1]]
    return sorted(courses)

def mode2_uniform_random(courses):
    random.shuffle(courses)
    return courses

def mode3_uniform_distribution(courses):
    gradation = []
    index = 0
    sublist = []
    grad = 2000
    for idx, c in enumerate(courses):
        match = re.findall(exp, c[0])
        courses[idx] = [match[0][1], match[0][0], c[1]]
    courses = sorted(courses)
    for c in courses:
        if int(c[0]) < grad:
            sublist.append(c)
        else:
            gradation.insert(index, sublist)
            sublist = []
            grad += 1000
            index += 1

    levels = len(gradation)
    transformed_courses = []
    while gradation != []:
        for i in range(levels):
            if gradation[i] != []:
                transformed_courses.append(gradation[i][0])
                gradation[i].pop(0)
            else:
                gradation.pop(i)
                levels -= 1
                break
    
    return transformed_courses
```

```python
def path_finder(semester_to_be_planned, schedule, courses_to_be_taken, pathfinding_mode):
    MAX_CREDITS = 18

    if pathfinding_mode == 1 or pathfinding_mode == 2:
        if pathfinding_mode == 1:
            courses_to_be_taken = mode1_lowest_first(courses_to_be_taken)
            semester_credits = 0
            for c in courses_to_be_taken:
                if int(c[2]) + semester_credits <= MAX_CREDITS:
                    semester_credits += c[2]
                    schedule[semester_to_be_planned].append(c)
                else:
                    print('Max credits reached')
                    break
            
        
        elif pathfinding_mode == 2:
            courses_to_be_taken = mode2_uniform_random(courses_to_be_taken)
        
            semester_credits = 0
            for c in courses_to_be_taken:
                if int(c[1]) + semester_credits <= MAX_CREDITS:
                    semester_credits += c[1]
                    schedule[semester_to_be_planned].append(c)
                else:
                    print('Max credits reached')
                    break
        
        

    elif pathfinding_mode == 3:
        classes = mode3_uniform_distribution(courses_to_be_taken)
        semester_credits = 0
        for c in classes:
            if int(c[2]) + semester_credits <= MAX_CREDITS:
                semester_credits += c[2]   
                schedule[semester_to_be_planned].append(c)
            else:
                print('Max credits reached')
                break
```

```python
schedule = {'summer 2026' : [], 'fall 2026': [], 'spring 2027' : []}
courses = extract_courses_from_html(HTML_PATH)
courses = [[c['course_code'], c['credits']] for c in courses]

path_finder('fall 2026', schedule, courses, 3)

schedule
```
